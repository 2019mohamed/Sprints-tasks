```python
import cv2 as cv
import matplotlib.pyplot as plt

i = cv.imread('Shades_framed_aa.jpg')
transi = cv.cvtColor(i, cv.COLOR_BGR2RGB)
plt.imshow(transi)
```

```python
b = cv.imread('Shades_framed_bayer.png',0)
b = cv.resize(b , (100,100))
plt.imshow(b)
```

```python
toi = cv.cvtColor(b , cv.COLOR_BAYER_GR2RGB)#bilinear Demosaicing. ref https://www.cse.iitb.ac.in/~ajitvr/CS663_Fall2016/demosaicing.pdf
#https://slazebni.cs.illinois.edu/spring19/assignment0.html
plt.imshow(toi)
```

```python
import torch.nn as nn
import torch
m = nn.Conv2d(1, 1,2,2)#convultion operator with random kernal
out = torch.tensor(b,dtype=torch.float).reshape(1,1,100,100)
out = m(out).reshape(50,50).detach()
plt.imshow(out)
```
