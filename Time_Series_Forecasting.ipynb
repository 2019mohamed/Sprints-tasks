```python
import numpy as np
import torch.nn as nn
import torch
from torch.utils.data import Dataset,DataLoader

def split_sequence(sequence, n_steps):
    x, y = list(), list()
    for i in range(len(sequence) - n_steps -1):
        end_ix = i + n_steps
        seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
        x.append(seq_x)
        y.append(seq_y)
    return np.array(x), np.array(y)
```

```python
n_steps = 12
size = int(len(data) * 0.7)
train, valid = data.iloc[0:size, -1], data.iloc[size:len(data), -1]
train_x,train_y = split_sequence(train.values, n_steps)
valid_x,valid_y = split_sequence(valid.values, n_steps)
```

```python
class TimeSeriesDataset(Dataset):
    def __init__(self,window,y):
        self.window = window
        self.y = y
    
    def __len__(self):
        return len(self.window)
    
    def __getitem__(self,idx):
        item = self.window[idx]
        label = self.y[idx]
        
        return item,label
```

```python
train = TimeSeriesDataset(train_x.reshape(train_x.shape[0],train_x.shape[1],1),train_y)
valid = TimeSeriesDataset(valid_x.reshape(valid_x.shape[0],valid_x.shape[1],1),valid_y)
train_loader = DataLoader(train,batch_size=1,shuffle=False, drop_last  = True)
valid_loader = DataLoader(valid,batch_size=1,shuffle=False , drop_last = True)
```

```python
class Model(nn.Module):
    def __init__(self):
        super(Model,self).__init__()
        self.conv1d = nn.Conv1d(12,128, kernel_size=1)
        self.relu = nn.ReLU()
        self.fc1 = nn.Linear(128 * 1 ,32)
        self.fc2 = nn.Linear(32,1)
        
    def forward(self,x):
      x = self.conv1d(x)
      x = self.relu(x)
      x = x.view(-1)
      x = self.fc1(x)
      x = self.relu(x)
      x = self.fc2(x)
        
      return x
```

```python
model = Model()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
criterion = nn.MSELoss()
```

```python
train_losses = []
valid_losses = []
def Train():
    
    running_loss = 0
    
    model.train()
    
    for idx, (inputs,labels) in enumerate(train_loader):
        inputs = inputs
        labels = labels
        optimizer.zero_grad()
        preds = model(inputs.float())
        loss = criterion(preds.float(),labels.float())
        loss.backward()
        optimizer.step()
        running_loss += loss
        
    train_loss = running_loss/len(train_loader)
    train_losses.append(train_loss.detach().numpy())
    
    print(f'train_loss {train_loss}')
    
def Valid():
    running_loss = 0
    
    model.eval()
    p = []
    with torch.no_grad():
        for idx, (inputs, labels) in enumerate(valid_loader):
            inputs = inputs
            labels = labels
            optimizer.zero_grad()
            preds = model(inputs.float())
            #p.append(preds.detach().numpy())
            #print(preds)
            #print(labels)
            loss = criterion(preds.float(),labels.float())
            running_loss += loss
            
        valid_loss = running_loss/len(valid_loader)
        valid_losses.append(valid_loss.detach().numpy())
        print(f'valid_loss {valid_loss}')
    #return p
```

```python
epochs = 50
for epoch in range(epochs):
  if epoch % 5 == 0:
    print('epochs {}/{}'.format(epoch+1,epochs))
    Train()
    Valid()
```

```python
from math import sqrt
train_losses = np.array(train_losses)
valid_losses = np.array(valid_losses)
print(train_losses.mean(),' ',sqrt(train_losses.mean()) ,' ',valid_losses.mean(),' ',sqrt(valid_losses.mean()))
```
